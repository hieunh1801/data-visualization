```python
from sklearn.linear_model import LinearRegression
```

```python
x1=np.array(x1).reshape((-1,1))
```

```python
y1=np.array(y1).reshape((-1,1))
```

```python
model=LinearRegression().fit(x1,y1)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(x1,y1)
```

```python
import statsmodels.api as sm
```

```python
model = sm.OLS(y1, x1)
```

```python
results = model.fit()
```

```python
print(results.summary())
```
