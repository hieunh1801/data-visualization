# Cars93 Data

```python
cars93=pd.read_csv('Datasets/Cars93.csv')
```

```python
car_size=cars93['Type']
car_manufacturer = cars93['Manufacturer']
```

```python
value, frequency = np.unique(car_size, return_counts=True)
value, frequency = np.unique(car_manufacturer, return_counts=True)
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(frequency, labels=value, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

```python
n=len(value)
theta = np.linspace(0.0, 2 * np.pi, n, endpoint=False)
radii = frequency
width = (2*np.pi)/n
ax = plt.subplot(111, polar=True)
ax.bar(theta, radii, width=width)
plt.show()
```

```python
# Create the pandas DataFrame 
df = pd.DataFrame({'Type': value, 'Frequency': frequency})
```

```python
df=df.sort_values('Frequency',ascending=False)
```

```python
value=df['Type']
frequency=df['Frequency']
n=len(value)
theta = np.linspace(0.0, 2 * np.pi, n, endpoint=False)
radii = frequency
width = (2*np.pi)/n
ax = plt.subplot(111, polar=True)
ax.bar(theta, radii, width=width)
plt.show()
```
