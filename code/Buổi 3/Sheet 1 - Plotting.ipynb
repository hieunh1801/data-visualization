# Vẽ các biểu đồ thể hiện thông tin cho bảng sau

```python
%matplotlib inline
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
# Step 1: Read File
df = pd.read_excel('datasets/data.xlsx', sheetname='Sheet1')
df
```

## Nhận xét
- Thông tin cần thể hiện ra:
    - 1 - Ứng với mỗi Domain gồm  cái là producer và consumer
    - 2 - Tổng là 100%
    - Cái lĩnh vực nào nhiều nhất (sort)
 

```python
# Biểu đồ 1
df.plot.bar(x='Domain', y=['Consumer', 'Producer'],color = ["red", "blue"])
# Thiết hụt: ko thể hiện được tổng giá trị là 100%
```

```python
# Biểu đồ 2: Biểu đồ đường 
df.plot()
# Chả hiển thị được cái gì hữu ích
```

```python
# Biểu đồ 3: bar - stack: cũng chỉ hiển thị được 2 trường dữ liệu - Thiếu cái 100%
df.plot.bar(stacked=True, x='Domain', y=['Consumer', 'Producer'],color = ["red", "blue"])
```
